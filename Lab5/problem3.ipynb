```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.neighbors import KNeighborsClassifier as KNN

# Suppress annoying pandas warnings
pd.options.mode.chained_assignment = None

data = pd.read_csv('/Users/anthonydepalatis/Documents/School/EE379K/Lab5/Data/Weekly.csv', 
                   usecols=['Year', 'Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume', 'Today', 'Direction'])
labels = pd.DataFrame(data['Direction'])

''' Part a '''
# Numerical Summary
print 'Part a: ' + "\n" + 'Numerical Summary -\n' + str(data.describe()) + "\n"

# Graphical Summary
#print 'Graphical Summary -\n'
sns.pairplot(data.drop('Direction', axis=1))
#plt.show()

''' Part b ''' # not sure if showing results correctly
# Perform Logistic Regression and print the results
b_data = data.drop(['Year', 'Direction', 'Today'], axis=1)
clf1 = LogisticRegression()
clf1.fit(b_data, labels)
#print 'Part b:\n ' + str(pd.DataFrame(clf1.transform(b_data)).describe()) + "\n"
print 'Part b:\n ' + str(b_data.describe()) + "\n"

''' Part c '''
confusion_matrix1 = confusion_matrix(labels, clf1.predict(b_data))
print 'Part c: ' + "\n" + str(confusion_matrix1)
print '% Correct Predictions: ' + str(float((confusion_matrix1[0][0] + confusion_matrix1[1][1])) / len(b_data) * 100) + '%\n'

''' Part d ''' 
# Trim the data to only be the years 1990 to 2008 Lag2 values
d_data = data[data['Year'] <= 2008]
d_labels = d_data['Direction']
d_data = d_data['Lag2']
clf2 = LogisticRegression()
clf2.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)

# Compute the confusion matrix for the remaining data
test_data = data[data['Year'] > 2008]
test_labels = pd.DataFrame(test_data['Direction'])
test_data = np.matrix(test_data['Lag2']).reshape(104, 1)
confusion_matrix2 = confusion_matrix(test_labels, clf2.predict(test_data))
print 'Part d: ' + "\n" + str(confusion_matrix2)
print '% Correct Predictions: ' + str(float((confusion_matrix2[0][0] + confusion_matrix2[1][1])) / len(test_data) * 100) + '%\n'

''' Part e '''
clf3 = LDA()
clf3.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)
confusion_matrix3 = confusion_matrix(test_labels, clf3.predict(test_data))
print 'Part e: ' + "\n" + str(confusion_matrix3)
print '% Correct Predictions: ' + str(float((confusion_matrix3[0][0] + confusion_matrix3[1][1])) / len(test_data) * 100) + '%\n'

''' Part f '''
clf4 = QDA()
clf4.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)
confusion_matrix4 = confusion_matrix(test_labels, clf4.predict(test_data))
print 'Part f: ' + "\n" + str(confusion_matrix4)
print '% Correct Predictions: ' + str(float((confusion_matrix4[0][0] + confusion_matrix4[1][1])) / len(test_data) * 100) + '%\n'

''' Part g '''
clf5 = KNN(n_neighbors=1)
clf5.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)
confusion_matrix5 = confusion_matrix(test_labels, clf5.predict(test_data))
print 'Part g: ' + "\n" + str(confusion_matrix5)
print '% Correct Predictions: ' + str(float((confusion_matrix5[0][0] + confusion_matrix5[1][1])) / len(test_data) * 100) + '%\n'

''' Part i '''
# KNN with k = 10
clf6 = KNN(n_neighbors=10)
clf6.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)
confusion_matrix6 = confusion_matrix(test_labels, clf6.predict(test_data))
print 'Part i: ' + "\n" + 'KNN with k = 10:\n' + str(confusion_matrix6)
print '% Correct Predictions: ' + str(float((confusion_matrix6[0][0] + confusion_matrix6[1][1])) / len(test_data) * 100) + '%\n'

# KNN with k = 100
clf7 = KNN(n_neighbors=100)
clf7.fit(np.asmatrix(d_data).reshape(985, 1), d_labels)
confusion_matrix7 = confusion_matrix(test_labels, clf7.predict(test_data))
print 'KNN with k = 100:\n' + str(confusion_matrix7)
print '% Correct Predictions: ' + str(float((confusion_matrix7[0][0] + confusion_matrix7[1][1])) / len(test_data) * 100) + '%\n'
```
